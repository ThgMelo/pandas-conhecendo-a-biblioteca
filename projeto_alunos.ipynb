# Descrição do desafio
1. Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.


2. Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.


3. Aplique um filtro que selecione apenas os alunos que foram aprovados.


4. Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".


## Importando os dados

```python
import pandas as pd
```

```python
pd.read_csv('alunos.csv')
```

```python
dados = pd.read_csv('alunos.csv')
dados
```

## Tratamento dos dados

```python
dados.isnull()
```

```python
dados.isnull().sum()
```

```python
dados.info()
```

```python
'''
Temos 6 dados nulos na coluna Notas e essa coluna possuí dados do tipo float.
Dessa forma, a melhor maneira de resolver é substituir os dados nulos por zero
'''
dados = dados.fillna(0)
```

```python
dados.isnull().sum()
```

## Removendo registros

```python
dados.query('Nome == "Carlos" | Nome == "Alice"')
```

```python
dados.query('Nome == "Carlos" | Nome == "Alice"').index
```

```python
registros_a_remover = dados.query('Nome == "Carlos" | Nome == "Alice"').index
```

```python
dados.drop(registros_a_remover, axis=0, inplace = True)
```

```python
dados.query('Nome == "Carlos" | Nome == "Alice"')
```

## Aplicando filtros
Aplique um filtro que selecione apenas os alunos que foram aprovados.

```python
dados.head()
```

```python
dados.Aprovado == True
```

```python
dados[dados.Aprovado == True]
```

```python
alunos_aprovados = dados[dados.Aprovado == True]
```

```python
alunos_aprovados
```

```python
alunos_aprovados.drop('Aprovado', axis = 1, inplace = True)
```

```python
alunos_aprovados.head()
```

## Salvando dados

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index = False)
```
